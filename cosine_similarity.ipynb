# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
X = [[0, 0, 0], [1, 1, 1]]
Y = [[1, 0, 0], [1, 1, 0]]
cosine_similarity(X, Y)
```

# Plot for Fun

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, Y)
plt.show()
```

```python
plt.plot(cosine_similarity(X, Y))
plt.show()
```
